```python
import csv
import random
import time
from flask import Flask, render_template,jsonify
from threading import Thread

app = Flask(__name__)

# 실시간으로 CSV 파일을 업데이트하는 함수
def update_csv():
    while True:
        # 임의의 데이터 생성
        x_value = time.strftime('%Y-%m-%d %H:%M:%S')
        y_values = [random.randint(0, 100) for _ in range(4)]  # 4개의 랜덤한 값을 생성

        # CSV 파일 업데이트
        with open('data.csv', 'a', newline='') as file:
            writer = csv.writer(file)
            writer.writerow([x_value] + y_values)  # x_value와 y_values를 함께 기록

        # 1초마다 업데이트
        time.sleep(1)

# Flask 애플리케이션 라우트
@app.route('/')
def index():
    return render_template('yap.html')

# CSV 파일을 읽어와서 데이터 반환
def get_data_from_csv():
    data = []
    with open('data.csv', 'r') as file:
        reader = csv.reader(file)
        next(reader)  # Skip the header row

        rows = list(reader)  # 모든 행을 리스트로 변환
        if len(rows) > 10:  # 최근 10개의 행만 유지
            rows = rows[-10:]

        for row in rows:
            if len(row) == 5:  # 열 개수가 5개인 경우에만 처리
                try:
                    data.append([row[0], int(row[1]), int(row[2]), int(row[3]), int(row[4])])
                except ValueError:
                    # Handle the case where the value is not a valid integer
                    print(f"Ignoring row with invalid value: {row}")
            else:
                print(f"Ignoring row with insufficient columns: {row}")
    return data
# 실시간 데이터를 반환하는 API 엔드포인트
@app.route('/api/data')
def get_realtime_data():
    data = get_data_from_csv()

    # 데이터를 적절한 형식으로 변환하여 반환
    labels = [row[0] for row in data]
    data1 = [row[1] for row in data]
    data2 = [row[2] for row in data]
    data3 = [row[3] for row in data]
    data4 = [row[4] for row in data]

    return jsonify({
        'labels': labels,
        'data1': data1,
        'data2': data2,
        'data3': data3,
        'data4': data4
    })

if __name__ == '__main__':
    # CSV 파일 생성
    with open('data.csv', 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['timestamp', 'value'])

    # CSV 파일 업데이트를 위한 스레드 실행
    update_thread = Thread(target=update_csv)
    update_thread.start()

    # Flask 애플리케이션 실행
    app.run()
```

```python
import csv
import random
import time
from flask import Flask, render_template, jsonify
from threading import Thread
import os

app = Flask(__name__,static_url_path='/static')

# 실시간으로 CSV 파일을 업데이트하는 함수
def update_csv():
    csv_path = os.path.join(app.root_path, 'static', 'data.csv')  # data.csv 파일의 경로 설정
    while True:
        # 임의의 데이터 생성
        y_values = [random.randint(0, 100) for _ in range(4)]  # 4개의 랜덤한 값을 생성

        # CSV 파일 업데이트
        with open(csv_path, 'a', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(y_values)  # y_values만 기록 (timestamp는 제외)

        # 1초마다 업데이트
        time.sleep(1)

# Flask 애플리케이션 라우트
@app.route('/')
def index():
    return render_template('nonreact_upgrade.html')

# CSV 파일을 읽어와서 데이터 반환
def get_data_from_csv():
    csv_path = os.path.join(app.root_path, 'static', 'data.csv')  # data.csv 파일의 경로 설정
    data = []
    with open(csv_path, 'r') as file:
        reader = csv.reader(file)
        for row in reader:
            if len(row) == 4:  # 열 개수가 4개인 경우에만 처리
                try:
                    data.append([int(value) for value in row])
                except ValueError:
                    # Handle the case where the value is not a valid integer
                    print(f"Ignoring row with invalid value: {row}")
            else:
                print(f"Ignoring row with insufficient columns: {row}")
    return data

# 실시간 데이터를 반환하는 API 엔드포인트
@app.route('/api/data')
def get_realtime_data():
    data = get_data_from_csv()

    # 데이터를 적절한 형식으로 변환하여 반환
    labels = list(range(len(data)))  # 인덱스를 레이블로 사용
    data1 = [row[0] for row in data]
    data2 = [row[1] for row in data]
    data3 = [row[2] for row in data]
    data4 = [row[3] for row in data]

    return jsonify({
        'labels': labels,
        'data1': data1,
        'data2': data2,
        'data3': data3,
        'data4': data4
    })

if __name__ == '__main__':
    # CSV 파일 생성
    with open('data.csv', 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['value1', 'value2', 'value3', 'value4'])

    # CSV 파일 업데이트를 위한 스레드 실행
    update_thread = Thread(target=update_csv)
    update_thread.start()

    # Flask 애플리케이션 실행
    app.run()
```

### html파일 중심 재코딩

```python
from flask import Flask, render_template
from flask_socketio import SocketIO
import pandas as pd
import threading

app = Flask(__name__)
app.config['SECRET_KEY'] = 'secret_key'
socketio = SocketIO(app, async_mode='eventlet')
thread = None

def read_csv_data():
    while True:
        data = pd.read_csv('data.csv')  # CSV 파일을 읽습니다.
        # 데이터 처리 및 그래프 업데이트 로직을 추가하세요.
        socketio.emit('update_chart', {'data': data.to_dict()}, namespace='/chart')  # 클라이언트에게 업데이트된 데이터 전송
        socketio.sleep(1)  # 1초마다 반복

@app.route('/')
def index():
    return render_template('index.html')

@socketio.on('connect', namespace='/chart')
def connect():
    global thread
    if thread is None:
        thread = socketio.start_background_task(read_csv_data)

if __name__ == '__main__':
    socketio.run(app, host='0.0.0.0', port=5000)
```
