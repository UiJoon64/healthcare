### 일상 [0,1,0] 레이블 데이터

```python
import os
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("./angle_data/daily/DDA_In_MIX_00385.csv")
df = df.drop(['Unnamed: 0'],axis=1)
input1 = df.values
input1.shape
```

```python
data_targets=[]
```

```python
#arr1 = np.ones((90,4))
arrays=[]
for direc in os.listdir("./angle_data/daily"):
    df = pd.read_csv("./angle_data/daily/{}".format(direc))
    df = df.drop(['Unnamed: 0'],axis=1)
    input1 = df.values
    arrays.append(input1)
    #arr1 = np.stack((arr1,df.values))
    print(input1.shape)
```

```python
# 배열을 (30, 4)의 shape로 분할하고 패딩
padded_arrays = []
for array in arrays:
    num_segments = array.shape[0] // 30
    remainder = array.shape[0] % 30
    
    if num_segments > 0:
        # (30, 4)로 분할
        segments = np.split(array[:num_segments*30], num_segments)
        padded_arrays.extend(segments)
    
    if remainder > 0:
        # 패딩된 (30, 4) 추가
        padding = np.tile(array[-1], (30 - remainder, 1))
        padded_array = np.concatenate([array[-remainder:], padding], axis=0)
        padded_arrays.append(padded_array)

# 결과 출력
for padded_array in padded_arrays:
    print(padded_array.shape)
```

```python
print(len(padded_arrays))
```

```python
# 분할된 input array 길이 측정
len_arr = len(padded_arrays)
len_arr
# target data 생성
tar_arr2 = []
for _ in range(len_arr):
    tar_arr2.append([0,1,0])
len(tar_arr2)
```

```python
for data,tar in zip(padded_arrays,tar_arr2):
    data_targets.append((data,tar))
```

### 데이터 증강

```python
import numpy as np
import random

# 데이터 증강 함수 정의
def augment_data(data, target, num_samples):
    augmented_data = []
    augmented_target = []
    
    for _ in range(num_samples):
        # 데이터 회전
        angle = random.uniform(0, 360)
        rotated_data = np.apply_along_axis(lambda x: np.roll(x, int(angle)), axis=0, arr=data)
        augmented_data.append(rotated_data)
        augmented_target.append(target)
        
        # 노이즈 추가
        noise = np.random.normal(0, 0.1, size=data.shape)
        noisy_data = data + noise
        augmented_data.append(noisy_data)
        augmented_target.append(target)
        
        # 시간 이동
        shift = random.randint(1, data.shape[0] - 1)
        shifted_data = np.roll(data, shift, axis=0)
        augmented_data.append(shifted_data)
        augmented_target.append(target)
        
        # 스케일 조정
        scale_factor = random.uniform(0.8, 1.2)
        scaled_data = data * scale_factor
        augmented_data.append(scaled_data)
        augmented_target.append(target)
    
    augmented_data = np.array(augmented_data)
    augmented_target = np.array(augmented_target)
    
    return augmented_data, augmented_target

# 데이터 증강 수행
num_samples = 10  # 추가할 데이터 샘플 수

augmented_data_targets = []
for data, target in data_targets:
    augmented_data, augmented_target = augment_data(data, target, num_samples)
    augmented_data_targets.extend(zip(augmented_data, augmented_target))

# 증강된 데이터 확인
print("원본 데이터 개수:", len(data_targets))
print("증강된 데이터 개수:", len(augmented_data_targets))
```

```python
data_targets = augmented_data_targets
data_targets
```

### 데이터 분할

```python
import random

# 전체 데이터를 훈련 및 검증 데이터로 분할하기 위한 비율
train_ratio = 0.8

# 데이터와 클래스(타겟) 쌍을 훈련 및 검증 데이터로 분할
random.shuffle(data_targets)  # 데이터 순서를 무작위로 섞음

# 분할 지점 계산
split_point = int(len(data_targets) * train_ratio)

# 훈련 데이터와 검증 데이터로 분할
train_data_targets = data_targets[:split_point]
val_data_targets = data_targets[split_point:]
```

```python
len(train_data_targets)
```

```python
len(val_data_targets)
```

### 모델 생성

```python
from tensorflow import keras
from tensorflow.keras.callbacks import EarlyStopping
import numpy as np

# LSTM 모델 구성
model = keras.Sequential()
model.add(keras.layers.LSTM(units=128, input_shape=(5, 4), return_sequences=True))
model.add(keras.layers.Dropout(0.2))  # Dropout 추가
model.add(keras.layers.LSTM(units=128))
model.add(keras.layers.Dropout(0.2))  # Dropout 추가
model.add(keras.layers.Dense(units=64, activation='relu', kernel_regularizer=keras.regularizers.l2(0.01)))  # Regularization 추가
model.add(keras.layers.Dense(units=3, activation='softmax'))

# 모델 컴파일
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# 데이터 전처리: 데이터와 클래스(타겟) 분리
train_data, train_targets = zip(*train_data_targets)
val_data, val_targets = zip(*val_data_targets)

# 데이터 전처리: numpy 배열로 변환
train_data = np.array(train_data)
train_targets = np.array(train_targets)
val_data = np.array(val_data)
val_targets = np.array(val_targets)

early_stopping = EarlyStopping(patience=10, restore_best_weights=True)

# 모델 학습
model.fit(train_data, train_targets, validation_data=(val_data, val_targets),
          epochs=100, batch_size=32, callbacks=[early_stopping])
```
